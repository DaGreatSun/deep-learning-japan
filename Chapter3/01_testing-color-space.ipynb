Lane Detection
===
Execution of Lane Detection via Image Pre-processing and Information Extraction

```python
import cv2
import os
import numpy as np

import sys
sys.path.append('../')

import general_utils
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [18, 6]
```

Image Import
===
3 files in total (sd1, sd2, sd3)

```python
original_images = []
file_paths = ["road-images/sd1.jpg","road-images/sd2.jpg","road-images/sd3.jpg"]
for file_path in file_paths:
    original_images.append(cv2.imread(file_path))

print_ = general_utils.mathplot_display_concatenated("Original Images", original_images, "h")
```

Color Space Testing
===
To Select best Color Space for Information Extraction

```python
images = original_images.copy()

# BGR Extraction
blue_channel = [image[:,:,0] for image in images]
print_ = general_utils.mathplot_display_concatenated("Extract Blue", blue_channel, "h")

green_channel = [image[:,:,1] for image in images]
print_ = general_utils.mathplot_display_concatenated("Extract Green", green_channel, "h")

red_channel = [image[:,:,2] for image in images]
print_ = general_utils.mathplot_display_concatenated("Extract Red", red_channel, "h")

# HSV Conversion
hsv_images = [cv2.cvtColor(image, cv2.COLOR_BGR2HSV) for image in images]
print_ = general_utils.mathplot_display_concatenated("HSV Conversion", hsv_images, "h")

# HLS Conversion
hls_images = [cv2.cvtColor(image, cv2.COLOR_BGR2HLS) for image in images]

hue_images = [image[:,:,0] for image in hls_images]
print_ = general_utils.mathplot_display_concatenated("Hue", hue_images, "h")

lightness_images = [image[:,:,1] for image in hls_images]
print_ = general_utils.mathplot_display_concatenated("Lightness", lightness_images, "h")

saturation_images = [image[:,:,2] for image in hls_images]
print_ = general_utils.mathplot_display_concatenated("Saturation", saturation_images, "h")
```
